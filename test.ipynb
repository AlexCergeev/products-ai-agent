```python
from langchain_gigachat.chat_models import GigaChat
from langchain_core.messages import HumanMessage
from _config import credentials
```

```python
giga = GigaChat(
    model="GigaChat",  
    credentials=credentials,
    # scope=<SCOPE>,
    verify_ssl_certs=False,
    temperature=1,
    top_p=0.5,
    maxTokens=100,
    profanity_check=True
)

resp = giga.invoke(
    [
        (
            "system",
            "Выступи в роли опытного руководителя IT продуктов. Решай задачу шаг за шагом.",
        ),
        HumanMessage(
            content='Напиши подробные требования для создания приложения для инвестирования. Глубоко проработай каждый пункт взаимодкйствия с клинета с приложением.',
            additional_kwargs={
                'attachments': []
            }
        ),
    ]
)
print(resp)
```
