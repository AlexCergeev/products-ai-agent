```python
from langchain_gigachat.embeddings import GigaChatEmbeddings

embeddings = GigaChatEmbeddings(credentials=credentials, verify_ssl_certs=False)
result = embeddings.embed_documents(texts=tz)
result2 = embeddings.embed_documents(texts=cd)
# result = embeddings.embed_query(text="Привет!")
# print(result)
```

```python
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Пример эмбеддингов (каждая строка — вектор для предложения)
embeddings1 = np.array(result)
embeddings2 = np.array(result2)

# Соответствующие предложения для каждого эмбеддинга
sentences1 = tz

sentences2 = cd

# Список для хранения найденных пар индексов
results = []

for i, vec in enumerate(embeddings1[:-1]):
    # Вычисляем косинусное сходство для каждого вектора из второго массива
    similarities = np.array([cosine_similarity(vec, vec2) for vec2 in embeddings2])
    sorted_idx = np.argsort(similarities)
    top_1, top_2, top_3 = sorted_idx[-1], sorted_idx[-2], sorted_idx[-3]
    results.append((i, (top_1, top_2, top_3)))
    
    # Выводим кортеж с индексами и соответствующие предложения
    print(f"Tuple: ({i}, {(top_1, top_2, top_3)})")
    print(f"Sentence 1: {sentences1[i]}")
    print(f"Sentence 2_top1: {sentences2[top_1]}")
    print(f"Sentence 2_top2: {sentences2[top_2]}")
    print(f"Sentence 2_top3: {sentences2[top_3]}")
    print("-" * 40)
```
